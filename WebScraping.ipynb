# Homework 4 question 2.
This returns a list of movies with ratings and names/year information that are alll rated above 8.5. It uses Beautifulsoup to get information from the page and then a single for loop to retrieve all information inside the title and rating columns

```python
import urllib.request

from bs4 import BeautifulSoup

baseUrl = "https://www.imdb.com/chart/top"
page = urllib.request.urlopen(baseUrl).read()
soup = BeautifulSoup(page, "html.parser") #Turn the url of the ranking page into a soup


ratings = soup.find_all(class_="ratingColumn imdbRating") #retrieve every column class dedicated to a movie's ranking
titles = soup.find_all(class_="titleColumn") #retrieve every column class dedicated to a movie's name/title
i = 0
print("IMDB Movies rated above 8.5: \n")
for title in titles: #Loop through the titles column to get info for each title
    name = title.find_all("a")  #Find the tag that the text of the title is kept in
    year = title.find_all("span") #Find the tag that the year text is kept in

    #Retrieving the text inside the tags
    movieYear = year[0].text 
    movieTitle = name[0].text
    #Find the tag that the rating is kept in and retrieve its text for the current movie
    allRatings = ratings[i].find_all("strong")
    movieRating = allRatings[0].text

    if(float(movieRating) > 8.5): #Check the rating text
        print(movieTitle, movieYear)
        print("Rated at ", movieRating, "\n\n")

    i += 1 #Iterate the index used to go through the movieRatings without using a nested for loop
```

# Homework 4 Question 3
This code also uses Beauifulsoup to turn the main url page into a soup. It then creates a separate list for headers, meant to be zipped with regarding the rows that each header would be associated with. It uses a a for loop to find the complete row information and create a dictionary. A nested for loop is then to populate the dictionary with each header/row value. The dictionary is then put into a dataframe.

```python
import urllib.request
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd

baseUrl = "https://www.worldometers.info/coronavirus/"
req = Request(baseUrl,headers={'User-Agent': 'Mozilla/5.0'})
page = urlopen(req).read()
soup = BeautifulSoup(page, "html.parser") #Turn the url of the ranking page into a soup


allCovidData=[]

#Create variables for headers of tables, rows, and columns
headers = []
table = soup.find("table") 
columns = table.find_all("th")
rows = table.tbody.find_all("tr") 

#Create a list of the headers used in the dataframe
i = 0
for th in columns:
    i = i + 1
    if((i>1) and (i < 7)):
        headers.append(th.text)
    else:
        continue
headers[0] = 'CountryName'


#Iterate through every row in covid info table
j = 0
for tr in table.tbody.find_all("tr"):
        rowHead = tr.find_all("td")
        
        
        headerList = list(zip(rowHead[1:], headers)) #Zip the rows with the headers and remove the # column 
        covidData={} #Create the dictionary to populate each row with covid info

#Populate the dictionary with Information next to each header
        for td, th in headerList: 
             covidData[th] = td.text.replace('\n', '')
        allCovidData.append(covidData)

#Create the dataframe
df = pd.DataFrame(allCovidData, columns=headers)

df = df.iloc[8: , :]  #Remove non-country information
df.head()
```
